## Detik

### Politik

```python
def scrape_detik(hal):
    global hades
    a = 1
    for page in range(1,hal):
        url = f'https://www.detik.com/search/searchall?query=politik&siteid=2'
        # url = f'https://www.detik.com/search/searchall?query=politik&sortby=time&page={page}'
        ge = req.get(url,hades).text #req: request yg di import sebelumnya
        sop = bs(ge,'lxml') #var yg menyimpan data dari bs
        li = sop.find('div',class_='list media_rows list-berita') #list berita dari web detik
        lin = li.find_all('article') #var untuk menyimpan artikel di tiap list
        for x in lin:
            link = x.find('a')['href']
            ge_ = req.get(link,hades).text
            sop_ = bs(ge_,'lxml')
            content = sop_.find_all('div',class_='detail__body-text itp_bodycontent')
            for x in content:
                x = x.find_all('p')
                y  = [y.text for y in x ]
                content_ = ''.join(y).replace('\n', '').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')
                a += 1
                with open('pol1.csv','a')as file:
                    wr = csv.writer(file, delimiter=',')
                    wr.writerow([content_])
```

```python
def scrape_cnbc(hal):
    for page in range(1, hal):
        url_cnbc = f'https://www.cnbcindonesia.com/search?query=politik&date='
        response = req.get(url_cnbc)
        soup = bs(response.text, 'html.parser')
        article_list = soup.find('ul', class_='list media_rows middle thumb terbaru gtm_indeks_feed')
        if article_list:
            articles = article_list.find_all('li')
            for article in articles:
                link_art = article.find('article').find('a')['href']
                response_art = req.get(link_art)
                soup_art = bs(response_art.text, 'html.parser')
                art = soup_art.find('div', class_='lm_content mt10')
                if art:
                    content_ = ''
                    warp = art.find_all('div', class_='detail_text')
                    for paragraph in warp:
                        for p in paragraph.find_all('p'):
                            content_ += p.text.strip() + ' '
                    with open('pol2.csv', 'a', newline='') as file:
                        writer = csv.writer(file)
                        writer.writerow([content_])
```

## Cleansing

Data cleansing is a preprocessing step in text mining which we remove any inaccurate, incomplete, inconsistent, or irrelevant data within a dataset. We want the data used in further prediction to be accurate, consistent, and reliable.

### Case folding

The data that we have is not structured and consistent, it is mixed between uppercase and lowercase. Hence, we will make sure that every letter is in the same term. In this case, I did a case folding and change all letters to lowercase.

```python
df = data.copy()
```

```python
df['Teks'] = df['Teks'].str.lower()
df
```

```python
def casefold_if_str(text):
    if isinstance(text, str):
        return text.casefold()
    else:
        return text

df['Teks'] = df['Teks'].apply(casefold_if_str)

df
```

### Remove Number

Because the presence of number and its value is not used in this analysis, we will remove all the number in the dataset.

```python
df['Teks'] = df['Teks'].replace(r'\b\d+\b', '', regex=True)
df
```

### Remove symbol and punctuation

Symbol and punctuation are irrelevant in this analysis and ought to be deleted from the dataset.

```python
import string
punctuation = '[' + string.punctuation + ']'
df['Teks'] = df['Teks'].replace(punctuation, ' ', regex=True)
df
```

```python
# Make sure that all non-string are removed
df['Teks'] = df['Teks'].apply(lambda x: str(x) if isinstance(x, str) else '')
df
```

### Remove double white space

The removed number, symbol, and punctuation are replaced by a white spice. So, in this step I will remove all of the double white space.

```python
df['Teks'] = df['Teks'].str.replace(r'\s{2,}', ' ', regex=True)
df
```

## Filtering

Common words that usually appear and have no meaning will be remove in this step. Removing unimportant word can reduce index size and processing time. Apart from that, it can also reduce noise.

### Remove stopwords

Stopwords are words that usually appear. Some example of stopwords are 'dan', 'lalu', 'jika', etc. These words are removed because it isn't important and doesn't have any context.

```python
import nltk
nltk.download('stopwords')
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
sw_id = set(stopwords.words('indonesian'))
df['Teks'] = df['Teks'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in sw_id]))
df
```

### Remove specific phrases

Besides stopwords, all the news from cnbc is started with 'jakarta cnbc indonesia'. This phrases is not important, thus will be removed.

```python
df['Teks'] = df['Teks'].str.replace('jakarta cnbc indonesia', '', regex=True)
df
```

##  Metode kedua: Word2Vec (skipgram)

This method represents words into a set of vectors that describe the usage in context. Vector size is 50 and the word frequency taken into account is at least 3.

Before creating a skipgram vector, tokenization will be done first.

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
```

```python
x_train_token =[word_tokenize(i) for i in x_train]
x_train_token
```

```python
x_test_token =[word_tokenize(i) for i in x_test]
x_test_token
```

After tokenizing, we will continue and make a skipgram model.

```python
import gensim
from gensim.models import Word2Vec
```

```python
model_skipgram = gensim.models.Word2Vec(x_train_token, min_count = 3,vector_size = 50, window = 5, sg=1)
```

```python
vocabulary_skipgram = model_skipgram.wv.index_to_key
print(vocabulary_skipgram)
```

```python
model_skipgram.wv.similarity('politik', 'negara')
```

```python
model_skipgram.wv.similarity('atlet', 'sepakbola')
```

```python
model_skipgram.wv.similarity('atlet', 'perusahaan')
```

```python
model_skipgram.wv.similarity('harimau', 'bisnis')
```

From the similarity test we can see that 'atlet' and 'sepakbola' have a bigger similarity than 'atlet' and 'perusahaan', which does make sense.

```python
model_skipgram.wv['atlet']
```

```python
x_train_skipgram = np.array([np.mean([model_skipgram.wv[word] for word in words if word in model_skipgram.wv] or [np.zeros(50)], axis = 0) for words in x_train_token])
x_train_skipgram
```

```python
x_test_skipgram = np.array([np.mean([model_skipgram.wv[word] for word in words if word in model_skipgram.wv] or [np.zeros(50)], axis = 0) for words in x_test_token])
x_test_skipgram
```

```python
print(len(x_train_skipgram))
```

The length of x_train_skipgram is the same as the x_train data we split earlier.
